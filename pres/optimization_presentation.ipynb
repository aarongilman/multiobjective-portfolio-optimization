### 1. Objectives in general

There are six objectives for portfolio:

1. Maximize expected return.
2. Maximize environmental, social, and corporate- sustainability.
3. Maximize dividend yield.
4. Maximize clean energy use.
5. Minimize price-earnings ratio.
6. Optimize volatility.

### 2. Decision variables

Using **proportional amount** of investment into a company $i$ as decision variable $x_i$.
 
In total 361 companies.

### 3. Objective functions

The objective functions could be then modelled as
$$
\begin{align}
 \text{maximize } &\{\sum_{i=1}^{n} r_ix_i,\sum_{i=1}^{n} s_ix_i,\sum_{i=1}^{n} d_ix_i,\sum_{i=1}^{n} c_ix_i,-\sum_{i=1}^{n} p_ix_i\}\\
 \text{optimize } &\sum_{i=1}^{n} \beta_ix_i\\
\end{align}
$$

where $r_i, s_i, d_i, c_i, p_i, \beta_i$ are expected return for investment, ESG-sustainability, dividend yield, clean energy use, price-earnings-ratio, and beta for company $i$, respectively.

1. Variable for expected return, $r_i$, is calculated as
$$
\begin{align}
r_i &= R_{f_m} + \beta_i(ER_m-R_{f_m}),\\
\text{where}\\
R_{f_m} &= \text{ the risk free-rate for market m},\\
\beta_i &= \text{ beta coefficient for company i},\\
ER_m &= \text{ expected average return on market m}.\\
\end{align}\\
$$
2. For $s_i$(ESG-sustainability), $d_i$(dividend yield), and $p_i$(price-earnings), the values are those of corresponding attribute values in data for company $i$.
3. For $c_i$(clean energy use), variable value is calculated as a sum of attributes 'Clean200' and 'ScienceBasedTargets' for company $i$. 
4. For optimizing volatility, portfolio beta can be aimed to be equal to 1. In this case, the volatility of the portfolio is same as the market's.

### 4. Constraints

1. Sum of weights should be equal to one. 

2. For diversifying portfolio, weight of a single asset should not be more than some value, e.g. 0.1. 

$$
\begin{align}
&\sum_{i=1}^n x_i = 1\\
  &0 \leq x_i \leq 0.1, i = 1,..,n
\end{align}
$$

### 5. Potential approaches to solve the problem

I have done this far:

1. Transform objective
$$
\begin{align}
 \text{optimize } &\sum_{i=1}^{n} \beta_ix_i\\
\end{align}
$$
into a pair of inequality constraints, so that
$$
\begin{align}
 &\sum_{i=1}^{n} \beta_ix_i + \epsilon \geq 1\\
 &\sum_{i=1}^{n} \beta_ix_i - \epsilon \leq 1\\
\end{align}
$$
where $\epsilon$ is a small value, e.g. 0.1,  to relax the constraints.

2. Solving the problem with epsilon constraint method, using 1) expected return and 2) clean energy use as objectives.
